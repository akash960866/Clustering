```python
#####################  Theoretical   #######################
#QUE 1 What is unsupervised learning in the context of machine learning?

# Unsupervised learning is a type of machine learning where the model is trained on data without labeled outputs.
# Unlike supervised learning, where each data point has an input and a known output (label), unsupervised learning works with
# only input data and tries to find patterns, structures, or groupings within it.

# Key Concepts:
#    No labels provided.
#    The model tries to learn the underlying structure of the data.
#    Common techniques:
#    Clustering (e.g., K-Means, Hierarchical)
#    Dimensionality Reduction (e.g., PCA, t-SNE, Autoencoders)
#    Anomaly Detection

# QUE 2 How does K-Means clustering algorithm work

# K-Means clustering is an iterative algorithm that aims to partition a dataset into k distinct, non-overlapping subgroups (clusters).
# Here's a breakdown of how it works:

# 1. Initialization: The algorithm starts by randomly selecting k initial cluster centers (centroids) from the data points.
# 2. Assignment: Each data point is assigned to the nearest centroid based on a distance metric (commonly Euclidean distance). This forms k initial clusters.
# 3. Update: The centroids of the newly formed clusters are recalculated as the mean of all data points within each cluster.
# 4. Iteration: Steps 2 and 3 are repeated. Data points are reassigned to the nearest new centroids, and the centroids are updated again.
# 5. Convergence: The process continues until the cluster assignments no longer change significantly, or a maximum number of iterations is reached. At this point, the algorithm is said to have converged.

# In essence, K-Means works by repeatedly refining the cluster assignments and centroid positions until a stable configuration is achieved, where data points are grouped with their nearest cluster centers.

# QUE 3 Explain the concept of a dendrogram in hierarchical clustering

# A dendrogram is a tree-like diagram that visualizes the steps involved in hierarchical clustering. It shows the groupings of data points at different levels of similarity or distance.

# Here's what a dendrogram typically illustrates:

# *   Leaves: Each leaf node at the bottom of the dendrogram represents a single data point.
# *   Branches: Branches connect data points or clusters that are merged together. The length of the branches represents the distance or dissimilarity between the merged clusters. Shorter branches indicate that the clusters are more similar and were merged earlier in the process.
# *   Nodes: Each node where branches merge represents a cluster formed by combining the data points or clusters below it.
# *   Height (or Y-axis): The height of the merge point on the dendrogram indicates the distance or dissimilarity at which the clusters were combined.

# By examining a dendrogram, you can:

# *   See how data points are grouped together at different levels of granularity.
# *   Determine the optimal number of clusters by visually inspecting the dendrogram for significant drops in height between merge points.
# *   Understand the relationships and similarities between different clusters.

# In essence, a dendrogram provides a visual history of the hierarchical clustering process, allowing you to explore the structure of the data and make informed decisions about cluster formation.

# QUE 4 What is the main difference between K-Means and Hierarchical Clustering

# The main difference between K-Means and Hierarchical Clustering lies in their approach to forming clusters:

# *   K-Means Clustering: This is a partitional clustering method. It aims to partition the data into a pre-specified number of k clusters. It's an iterative process where data points are assigned to the nearest centroid, and centroids are updated until convergence. K-Means requires you to specify the number of clusters (k) beforehand.

# *   Hierarchical Clustering: This is a hierarchical clustering method. It builds a hierarchy of clusters, either by starting with individual data points and merging them into larger clusters (agglomerative) or by starting with one large cluster and splitting it into smaller ones (divisive). Hierarchical clustering does not require you to specify the number of clusters beforehand. Instead, you can decide on the number of clusters by cutting the dendrogram at a certain height.

# In summary, K-Means creates a flat partition of the data into a fixed number of clusters, while Hierarchical Clustering creates a tree-like structure of clusters, allowing you to explore relationships at different levels of granularity.

# QUE 5 What are the advantages of DBSCAN over K-Means?

# DBSCAN (Density-Based Spatial Clustering of Applications with Noise) has several advantages over K-Means:
# *   Handles Arbitrary Shapes: K-Means assumes clusters are spherical, whereas DBSCAN can find clusters of arbitrary shapes.
# *   Robust to Noise: DBSCAN can identify and separate noise points (outliers) that don't belong to any cluster, which K-Means struggles with.
# *   Doesn't Require Number of Clusters: K-Means requires you to specify the number of clusters (k) beforehand, which can be difficult to determine. DBSCAN, on the other hand, determines the number of clusters based on the density of the data.

# QUE 6 When would you use Silhouette Score in clustering?

# The Silhouette Score is a metric used to evaluate the quality of clusters produced by clustering algorithms like K-Means or hierarchical clustering. You would use it:
# *   To determine the optimal number of clusters: By calculating the Silhouette Score for different numbers of clusters, you can choose the k that results in the highest score, indicating better-defined clusters.
# *   To compare different clustering algorithms: You can use the Silhouette Score to compare the performance of different clustering algorithms on the same dataset.
# *   To assess the quality of individual clusters: The Silhouette Score provides a score for each data point, indicating how well it fits into its assigned cluster compared to other clusters. This can help identify poorly clustered points.

# QUE 7 What are the limitations of Hierarchical Clustering?

# Limitations of Hierarchical Clustering:
# *   Computational Complexity: Hierarchical clustering can be computationally expensive, especially for large datasets, as it requires calculating and storing pairwise distances between data points.
# *   Sensitivity to Noise and Outliers: Outliers can significantly affect the merging process and lead to suboptimal clustering results.
# *   Difficulty with Irregular Shapes: Like K-Means, hierarchical clustering can struggle with clusters that are not well-separated or have irregular shapes.
# *   Choosing the Right Cut-off: Deciding where to cut the dendrogram to obtain the desired number of clusters can be subjective and may require domain knowledge or the use of evaluation metrics.

# QUE 8 Why is feature scaling important in clustering algorithms like K-Means?

# Why feature scaling is important in clustering algorithms like K-Means:
# Feature scaling is crucial for distance-based clustering algorithms like K-Means because they rely on calculating distances between data points to form clusters. If features have different scales (e.g., one feature is in meters and another is in kilometers), the feature with the larger scale will dominate the distance calculations, regardless of its actual importance. This can lead to biased clustering results where the algorithm is more influenced by features with larger values.

# Feature scaling ensures that all features contribute equally to the distance calculations by bringing them to a similar scale. Common scaling techniques include:
# *   Standardization (Z-score scaling): Transforms the data to have a mean of 0 and a standard deviation of 1.
# *   Normalization (Min-Max scaling): Scales the data to a fixed range, usually between 0 and 1.

# By scaling the features, you ensure that the clustering algorithm accurately captures the underlying structure of the data based on the true relationships between data points, rather than being skewed by differences in feature scales.

# QUE 9 How does DBSCAN identify noise points?

# DBSCAN identifies noise points as data points that are neither core points nor border points.

# *   Core points: A data point is a core point if there are at least min_samples (a parameter of DBSCAN) data points within a distance of eps (another parameter) from it (including the point itself).
# *   Border points: A data point is a border point if it is not a core point itself but is within the eps distance of a core point.
# *   Noise points: Any data point that does not meet the criteria to be a core point or a border point is considered a noise point. These points are not part of any cluster.

# QUE 10 Define inertia in the context of K-Means

# Inertia, also known as the within-cluster sum of squares (WCSS), is a metric used to evaluate the quality of a K-Means clustering. It is the sum of squared distances of samples to their closest cluster center.

# In simpler terms, inertia measures how "tightly" the data points are clustered around their respective centroids. A lower inertia value generally indicates better clustering, as it means the data points are closer to their cluster centers.

# However, inertia has a limitation: it tends to decrease as the number of clusters (k) increases. This is because with more clusters, the data points will naturally be closer to their assigned centroids. Therefore, inertia alone is not sufficient to determine the optimal number of clusters, and it's often used in conjunction with other methods like the elbow method.

# QUE 11 What is the elbow method in K-Means clustering?

# The elbow method is a heuristic used in K-Means clustering to help determine the optimal number of clusters (k).

# Here's how it works:

# 1. You run the K-Means algorithm for a range of k values (e.g., from 1 to 10).
# 2. For each k, you calculate the inertia (within-cluster sum of squares - WCSS). Inertia measures how close the data points are to their assigned cluster centers.
# 3. You plot the inertia values against the number of clusters (k).
# 4. The plot will typically show a rapid decrease in inertia as k increases initially. However, at some point, the rate of decrease will slow down significantly, forming an "elbow" shape in the plot.

# The idea behind the elbow method is that the "elbow" point represents the optimal number of clusters. Beyond this point, adding more clusters doesn't significantly reduce the inertia, suggesting that you are likely just splitting existing clusters rather than forming new, distinct ones.

# It's important to note that the elbow method is a heuristic and not a definitive rule. Sometimes the elbow is not clearly defined, and other metrics (like the Silhouette Score) or domain knowledge might be needed to make the final decision on the number of clusters.

# QUE 12 Describe the concept of "density" in DBSCAN

# In DBSCAN, the concept of "density" is central to how it defines and forms clusters. Instead of looking at distances between all points like in K-Means or hierarchical clustering, DBSCAN focuses on how many points are located within a certain radius of a given point.

# Here's a breakdown of the key ideas related to density in DBSCAN:

# *   Epsilon (eps): This parameter defines the maximum distance between two samples for one to be considered as in the neighborhood of the other. It's essentially the radius around a point that the algorithm checks for other points.
# *   Minimum Samples (min_samples): This parameter defines the number of samples in a neighborhood for a point to be considered as a core point.

# Based on these parameters, DBSCAN categorizes points into three types:

# *   Core Point: A point is a core point if its neighborhood (defined by eps) contains at least min_samples number of points (including the point itself). These are the "dense" points that form the core of a cluster.
# *   Border Point: A point is a border point if it is not a core point itself but is within the eps distance of a core point. These points are on the edge of a cluster.
# *   Noise Point: A point is a noise point if it is neither a core point nor a border point. These points are in low-density regions and are not considered part of any cluster.

# DBSCAN then forms clusters by connecting core points that are within eps distance of each other, and including border points that are in the neighborhood of these core points. Clusters are essentially maximal sets of density-reachable points.

# In essence, DBSCAN defines clusters as areas of high density separated by areas of low density. The eps and min_samples parameters control what is considered "high density".

# QUE 13 Can hierarchical clustering be used on categorical data?

# Hierarchical clustering can be used on categorical data, but it's not as straightforward as using it on numerical data. This is because most standard hierarchical clustering algorithms (like those using Euclidean distance) are designed for numerical data and rely on calculating distances or similarities between data points, which isn't directly applicable to categories.

# To use hierarchical clustering with categorical data, you need to convert the categorical data into a numerical format that allows for distance calculations. Common approaches include:

# 1.  One-Hot Encoding: Convert each category into a binary feature. After one-hot encoding, you can use distance metrics suitable for binary data, like Jaccard distance or Hamming distance.
# 2.  Using Appropriate Distance Metrics: Instead of Euclidean distance, you can use distance metrics specifically designed for categorical data or mixed data types, such as Gower distance, Hamming distance, or Jaccard distance.

# In summary, while you can't directly apply standard hierarchical clustering to raw categorical data, you can prepare your data (e.g., through encoding) and use appropriate distance metrics to make it compatible with hierarchical clustering algorithms.

# QUE 14 What does a negative Silhouette Score indicate?

# A negative Silhouette Score for a data point indicates that the data point might have been assigned to the wrong cluster.

# Here's a breakdown of what the Silhouette Score means:

# *   The Silhouette Score for a single data point measures how similar that data point is to its own cluster (cohesion) compared to other clusters (separation). The score ranges from -1 to 1.

# *   A score close to +1 means the data point is well within its own cluster and far from other clusters.
# *   A score around 0 means the data point is near the decision boundary between two neighboring clusters.
# *   A score close to -1 means the data point might have been assigned to the wrong cluster, as it is more similar to data points in another cluster than to those in its own cluster.

# Therefore, a negative Silhouette Score suggests that the data point is potentially misclassified and is more similar to a different cluster than the one it's currently in. When evaluating the overall quality of a clustering, the average Silhouette Score is often used. A higher average score indicates better clustering.

# QUE 15 Explain the term "linkage criteria" in hierarchical clustering

# In hierarchical clustering, "linkage criteria" (also known as linkage methods) determine how the distance between two clusters is calculated. When the algorithm merges two clusters, it needs a way to quantify how "far apart" those clusters are. The linkage criteria provide this rule.

# Here are some common linkage criteria:

# *   Single Linkage (Minimum Linkage): The distance between two clusters is the minimum distance between any single data point in one cluster and any single data point in the other cluster. This method tends to create long, chain-like clusters and is sensitive to noise and outliers.
# *   Complete Linkage (Maximum Linkage): The distance between two clusters is the maximum distance between any single data point in one cluster and any single data point in the other cluster. This method tends to create more compact, spherical clusters.
# *   Average Linkage: The distance between two clusters is the average distance between all pairs of data points, where one point is from each cluster. This method is a compromise between single and complete linkage and is less sensitive to outliers than single linkage.
# *   Centroid Linkage: The distance between two clusters is the distance between the centroids (mean of all data points) of the two clusters.
# *   Ward's Method: This method minimizes the variance of the clusters being merged. It tends to create clusters of roughly equal size.

# The choice of linkage criteria can significantly impact the resulting dendrogram and the structure of the clusters. The best choice often depends on the specific dataset and the desired cluster characteristics.

# QUE 16 Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

# K-Means works by finding cluster centers (centroids) and assigning points to the nearest centroid. It implicitly assumes that clusters are roughly spherical and of similar size and density. When clusters have very different sizes or densities:

# *   Varying Sizes: A large cluster might "attract" points from a smaller, nearby cluster, causing the smaller cluster to be absorbed or misrepresented.
# *   Varying Densities: In data with clusters of different densities, K-Means might split dense clusters or merge sparse ones, as it tries to minimize the overall within-cluster sum of squares, which is influenced by the spread of points.

# QUE 17 What are the core parameters in DBSCAN, and how do they influence clustering?

# DBSCAN has two main parameters:

# 1.  eps (epsilon): This defines the maximum distance between two samples for one to be considered as in the neighborhood of the other.
#     *   Influence: A smaller eps means only points very close to each other will be considered neighbors, leading to more numerous, smaller clusters or more points being classified as noise. A larger eps considers points further away as neighbors, potentially merging clusters or incorrectly including noise points in clusters.
# 2.  min_samples: This is the number of samples in a neighborhood for a point to be considered as a core point.
#     *   Influence: A higher min_samples makes the density requirement for a core point stricter. This can lead to fewer, larger clusters and more points being classified as noise. A lower min_samples makes it easier for points to be considered core points, potentially leading to more clusters and fewer noise points.

# These parameters are crucial for DBSCAN's ability to find density-connected clusters and identify outliers. Choosing appropriate values for eps and min_samples is key to effective DBSCAN clustering and often requires domain knowledge or experimentation.

# QUE 18 How does K-Means++ improve upon standard K-Means initialization?

# Standard K-Means initialization typically involves randomly selecting k data points as the initial centroids. While simple, this can lead to poor clustering results if the initial centroids are not well-distributed. If initial centroids are too close to each other or fall within the same dense cluster, the algorithm might converge to a suboptimal solution (a local minimum) where the clusters are not well-formed or accurately represent the underlying data structure.

# K-Means++ addresses this issue by using a smarter approach to select the initial centroids. The key idea is to select initial centroids that are spread out from each other. Here's how it works:

# 1.  Select the first centroid: One data point is chosen randomly from the dataset as the first centroid.
# 2.  Calculate distances: For each remaining data point, the squared Euclidean distance to the nearest already selected centroid is calculated.
# 3.  Select the next centroid: The next centroid is chosen from the remaining data points with a probability proportional to its squared distance from the nearest already selected centroid. This means that points that are farther away from existing centroids have a higher chance of being selected as the next centroid.
# 4.  Repeat: Steps 2 and 3 are repeated until k centroids have been selected.

# Why this is better: By selecting centroids that are progressively farther away from the ones already chosen, K-Means++ ensures that the initial centroids are more spread out across the data space. This reduces the likelihood of starting with centroids that are all clustered together in one area, which in turn increases the chances of the K-Means algorithm converging to a better, more globally optimal clustering solution. This leads to more accurate and stable clustering results.

# QUE 19 What is agglomerative clustering?

# Agglomerative clustering is a type of hierarchical clustering. It's a "bottom-up" approach where the algorithm starts with each data point as its own individual cluster and then iteratively merges the closest clusters until only one cluster (or a pre-defined number of clusters) remains.

# Here's a breakdown of the process:

# 1.  Initialization: Each data point is considered a single cluster. If you have n data points, you start with n clusters.
# 2.  Calculate Distances: The distances or similarities between all pairs of clusters are calculated. The "distance" between clusters is determined by the chosen "linkage criteria" (which we discussed earlier, e.g., single linkage, complete linkage, average linkage, Ward's method).
# 3.  Merge Closest Clusters: The two closest clusters (based on the linkage criteria) are merged into a single new cluster.
# 4.  Repeat: Steps 2 and 3 are repeated. The distances between the new set of clusters are calculated, and the two closest are merged. This process continues until only one cluster remains, containing all the data points.

# The result of agglomerative clustering is a hierarchy of clusters, which can be visualized as a dendrogram. By cutting the dendrogram at different heights, you can obtain different numbers of clusters.

# QUE 20 What makes Silhouette Score a better metric than just inertia for model evaluation?

# While inertia (within-cluster sum of squares) is a useful metric in K-Means for indicating how tightly data points are clustered around their centroids, it has a significant limitation: it always decreases as the number of clusters (k) increases. This is because with more clusters, the data points will naturally be closer to their assigned centroids, leading to a lower WCSS. This makes it difficult to determine the optimal number of clusters based on inertia alone, as it doesn't penalize for having too many clusters.

# The Silhouette Score, on the other hand, provides a more nuanced evaluation of clustering quality by considering both the cohesion within clusters and the separation between clusters. For each data point, it calculates a silhouette coefficient that measures how similar the data point is to its own cluster compared to other clusters.

# The Silhouette Score for a data point i is calculated as:

# `s(i) = (b(i) - a(i)) / max(a(i), b(i))`

# Where:

# *   `a(i)` is the average distance of data point i to all other data points in the same cluster (a measure of cohesion). A lower `a(i)` indicates better cohesion.
# *   `b(i)` is the average distance of data point i to all data points in the nearest other cluster (a measure of separation). A higher `b(i)` indicates better separation.

# The Silhouette Score for the entire clustering is the average of the silhouette coefficients for all data points. The score ranges from -1 to +1:

# *   Close to +1: Indicates that data points are well-clustered and clearly separated from other clusters.
# *   Close to 0: Indicates that data points are near the decision boundary between two neighboring clusters.
# *   Close to -1: Indicates that data points may have been assigned to the wrong cluster.

# Why Silhouette Score is better than inertia:

# *   Considers both cohesion and separation: Unlike inertia, which only considers within-cluster distances, the Silhouette Score takes into account how well-separated the clusters are from each other.
# *   Not monotonically decreasing with k: The Silhouette Score does not always decrease as k increases. It can help identify an optimal number of clusters where the clusters are both cohesive and well-separated.
# *   Provides insight into individual data points: The individual silhouette coefficients can help identify data points that are poorly clustered or might be outliers.

# In summary, while inertia is useful for the elbow method, the Silhouette Score provides a more comprehensive evaluation of clustering quality by balancing within-cluster cohesion and between-cluster separation, making it a better metric for comparing different clustering results and identifying the optimal number of clusters.
```

```python
# QUE 21 Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
#        scatter plot

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Step 1: Generate synthetic data
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)

# Step 2: Apply K-Means Clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Step 3: Visualize the Clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', s=200, marker='X', label='Centroids')
plt.title('K-Means Clustering with 4 Centers')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#QUE 23 Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot

from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Generate synthetic moon-shaped data
X, _ = make_moons(n_samples=300, noise=0.1, random_state=42)

# Step 2: Apply DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)

# Step 3: Visualize clusters and highlight outliers
plt.figure(figsize=(8, 6))
unique_labels = set(labels)

for label in unique_labels:
    if label == -1:
        # Outliers (noise points)
        color = 'red'
        marker = 'x'
        label_name = 'Outliers'
    else:
        color = plt.cm.Set1(label / 10)
        marker = 'o'
        label_name = f'Cluster {label}'

    plt.scatter(X[labels == label, 0], X[labels == label, 1],
                c=color, label=label_name, marker=marker, edgecolor='k', s=60)

plt.title('DBSCAN on Moons Dataset (Outliers Highlighted)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#QUE 27 Generate synthetic data using make_blobs with varying cluster standard deviations and cluster with
#        DBSCAN

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

# Step 1: Generate synthetic data with varying std deviations
centers = [[-5, -2], [0, 0], [5, 5]]
cluster_std = [0.8, 2.5, 0.5]  # Different spreads
X, y_true = make_blobs(n_samples=750, centers=centers, cluster_std=cluster_std, random_state=42)

# Step 2: Apply DBSCAN
db = DBSCAN(eps=0.9, min_samples=5)
labels = db.fit_predict(X)

# Step 3: Plot the result
plt.figure(figsize=(8, 6))
unique_labels = set(labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

for label, color in zip(unique_labels, colors):
    class_member_mask = (labels == label)
    if label == -1:
        # Outliers
        plt.plot(X[class_member_mask, 0], X[class_member_mask, 1], 'o',
                 markerfacecolor='k', markersize=5, label='Outlier')
    else:
        plt.plot(X[class_member_mask, 0], X[class_member_mask, 1], 'o',
                 markerfacecolor=tuple(color[:3]), markersize=7, label=f'Cluster {label}')

plt.title("DBSCAN Clustering on make_blobs Data with Varying Std Dev")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#QUE 29 Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Step 1: Generate synthetic data
X, _ = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

# Step 2: Evaluate silhouette scores for k = 2 to 5
silhouette_scores = []

k_values = range(2, 6)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

# Step 3: Plot silhouette scores as a bar chart
plt.figure(figsize=(8, 5))
plt.bar(k_values, silhouette_scores, color='skyblue')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for k = 2 to 5")
plt.xticks(k_values)
plt.ylim(0, 1)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#QUE 31 Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
#        decision boundaries

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Step 1: Generate synthetic data with overlapping clusters
X, y_true = make_blobs(n_samples=300,
                       centers=3,
                       cluster_std=2.0,  # Higher std for overlapping
                       random_state=42)

# Step 2: Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.predict(X)

# Step 3: Create meshgrid for decision boundaries
h = 0.1
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Step 4: Plot decision boundaries and clusters
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='Pastel1')  # Decision boundaries
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='Set1', edgecolor='k', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='black', marker='X', s=200, label='Centroids')
plt.title("K-Means Clustering with Decision Boundaries (Overlapping Blobs)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()

```

```python
# QUE 32 Load the Digits dataset and apply DBSCAN after reducing dimensions with t-SNE. Visualize the results

from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Load the Digits dataset
digits = load_digits()
X = digits.data

# Step 2: Reduce dimensions to 2D using t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X)

# Step 3: Apply DBSCAN
dbscan = DBSCAN(eps=5, min_samples=5)
labels = dbscan.fit_predict(X_tsne)

# Step 4: Visualize the results
plt.figure(figsize=(10, 6))
unique_labels = set(labels)

for label in unique_labels:
    if label == -1:
        # Outliers
        color = 'red'
        marker = 'x'
        label_name = 'Outliers'
    else:
        color = plt.cm.tab10(label % 10 / 10)
        marker = 'o'
        label_name = f'Cluster {label}'

    plt.scatter(X_tsne[labels == label, 0], X_tsne[labels == label, 1],
                c=color, marker=marker, label=label_name, edgecolor='k', s=60)

plt.title("DBSCAN on Digits Dataset (t-SNE Reduced)")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
#QUE 34  Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a
# line plot
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Step 1: Load Breast Cancer dataset
data = load_breast_cancer()
X = data.data

# Step 2: Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Compare inertia for K = 2 to 6
inertias = []
k_values = range(2, 7)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

# Step 4: Plot the inertia values
plt.figure(figsize=(8, 5))
plt.plot(k_values, inertias, marker='o', linestyle='-', color='teal')
plt.title("K-Means Inertia for K = 2 to 6 (Breast Cancer Dataset)")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia (Within-Cluster Sum of Squares)")
plt.grid(True)
plt.xticks(k_values)
plt.show()

```

```python
#QUE 44 Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Load the Iris dataset
data = load_iris()
X = data.data

# Step 2: Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
cluster_labels = kmeans.fit_predict(X)

# Step 3: Compute Silhouette Scores
silhouette_vals = silhouette_samples(X, cluster_labels)
avg_score = silhouette_score(X, cluster_labels)

# Step 4: Plot silhouette scores for each sample
plt.figure(figsize=(10, 6))
y_lower = 10
for i in np.unique(cluster_labels):
    ith_cluster_silhouette_vals = silhouette_vals[cluster_labels == i]
    ith_cluster_silhouette_vals.sort()
    size_cluster_i = ith_cluster_silhouette_vals.shape[0]
    y_upper = y_lower + size_cluster_i

    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      ith_cluster_silhouette_vals, alpha=0.7, label=f"Cluster {i}")

    y_lower = y_upper + 10  # add gap between clusters

# Step 5: Average line and formatting
plt.axvline(x=avg_score, color="red", linestyle="--", label=f"Avg Silhouette = {avg_score:.2f}")
plt.title("Silhouette Plot for KMeans Clustering (Iris Dataset)")
plt.xlabel("Silhouette Coefficient")
plt.ylabel("Samples")
plt.legend()
plt.grid(True)
plt.show()
```
